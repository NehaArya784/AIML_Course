### Mean Imputation in 'Age' Column

```python
#Calculate the mean of the 'Age' column

mean_Age=data['Age'].mean()
mean_Age
```

```python
#Create a copy of the data to avoid modifying the original dataset

data1=data.copy()
```

```python
#Fill missing values in 'Age' column with the mean value

data1['Age'].fillna(mean_Age)
```

```python
#For removing the decimal part from the data use floor division

mean_Age//1
```

```python
#Round the mean value to a 0 places

round(mean_Age)
```

```python
#Fill missing values in 'Age' column with the round mean value

data1['Age']=data1['Age'].fillna(round(mean_Age),inplace=True)
```

```python
data1
```
