## Histogram

```python
# import seaborn library

import seaborn as sns
```

```python
#Load the iris dataset 

iris=sns.load_dataset('iris')
```

```python
iris.head()   #print first 5 rows of the dataset
iris.info()   #check the data types and null values in the dataframe
iris.describe  #for representing the statistical data from the data
```

```python
sns.histplot(data=iris,x='petal_length',bins=100,kde=True,color='blue')
```

```python
sns.kdeplot(data=iris,x='petal_length',hue='species',color='red',fill=True,alpha=0.5)
```

```python
column=['species']

#Count of each species
iris['species'].value_counts()
```

```python
#Skewedness check

sns.countplot(x='species',data=iris)
```

## Pairplot

```python
sns.pairplot(iris,markers=["o","s","p"],hue='species',palette='husl')
```
